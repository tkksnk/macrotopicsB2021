```julia
using SparseArrays
using LinearAlgebra

# NOTE 210625: the case with δ=1.0 can be solved with implicit method only?

# parameters
σ = 2.0       # CRRA
α = 0.3       # capital share
δ = 1.0 # depreciation rate # 1.0 does not work
ρ = 0.05      # discount rate 
A = 1         # productivity

# steady state capital
kss = (α*A/(ρ+δ))^(1/(1-α))
println(kss)

# number of grid points
# nk = 150 # too many grid points sometimes do not work
nk = 10001 #150
# set up the grid
# too narrow bounds do not work (say [0.9,1.1]kss)
kmin = 0.001*kss
kmax = 2.0*kss
# evenly spaced grid
# k = collect(range(kmin,stop=kmax,length=nk))
# log spaced grid
k = collect(exp10.(range(log(kmin-kmin+1)/log(10),stop=log(kmax-kmin+1)/log(10),length=nk)))
k = k .+ (kmin-1.0)
# distance between each grid
# dk = (kmax-kmin)/(nk-1)

maxit = 10000 # maximum number of iterations
crit  = 1e-6; # convergence criteria
```

```julia
# INITIAL GUESS
v0 = (A.*k.^α).^(1.0-σ)/(1.0-σ)/ρ
# Delta = 0.0001 # damping parameter

# v,c = vfict_explicit(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta);
# v,c = vfict_explicit_nomatrix(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta);
Delta = 0.1 # δ=1.0 and Delta=0.2 does not work
v,c = vfict_implicit(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta);
```

```julia
function vfict_explicit_nomatrix(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta)

    nk = size(k,1)
    kmin = minimum(k)
    kmax = maximum(k)
#     dk = (kmax-kmin)/(nk-1)
    dkf = [k[2:nk]-k[1:nk-1]; k[nk]-k[nk-1]]
    dkb = [k[2]-k[1]; k[2:nk]-k[1:nk-1]]

    # for iteration
    dVf = zeros(nk) # forward difference
    dVb = zeros(nk) # backward difference
    c   = zeros(nk) # optimal consumption (from dV)

    v = v0 # new v
    V = v  # old v

    dist  = zeros(maxit); # distance between old and new value functions at each iteration    

    # MAIN LOOP
    for n in 1:maxit

        # forward difference
        dVf[1:nk-1] = (V[2:nk]-V[1:nk-1])./dkf[1:nk-1]
        dVf[nk] = 0 # will never be used (as s<0)
        # backward difference
        dVb[2:nk] = (V[2:nk]-V[1:nk-1])./dkb[2:nk]
        dVb[1] = 0 # will never be used (as s>0)

        # consumption and savings with forward difference
        cf = dVf.^(-1/σ)
        sf = A.*k.^α - δ.*k - cf
        # consumption and savings with backward difference
        cb = dVb.^(-1/σ)
        sb = A.*k.^α - δ.*k - cb
        # consumption and derivative of value function at steady state (used for c only)
        c0  = A.*k.^α - δ.*k
        dV0 = c0.^(-σ)

        # dV_upwind makes a choice of forward or backward differences based on
        # the sign of the drift    
        If = sf .> 0       # below steady state
        Ib = sb .< 0       # above steady state
        I0 = (1.0.-If.-Ib) # at steady state

        # make sure the right approximations are used at the boundaries
        Ib[1] = 0 
        If[1] = 1
        Ib[nk] = 1
        If[nk] = 0

        dV_Upwind = dVf.*If + dVb.*Ib + dV0.*I0 # important to include third term    
        c = dV_Upwind.^(-1.0/σ)
        u = c.^(1.0-σ)/(1.0-σ)

        # explicit method, without sparse matrix
        Vchange = u + dV_Upwind.*(A.*k.^α - δ.*k - c) - ρ.*V
    #     Vchange = u + (dVf.*If + dVb.*Ib).*(A.*k.^α - δ.*k - c) - ρ.*V
        v = V + Delta*Vchange # new v
        V = v

        dist[n] = maximum(abs.(Vchange))
        println([n,dist[n]])
        if dist[n]<crit
            println("Value Function Converged, Iteration = ")
            println(n)
            break
        end
    end
    
    return v, c
    
end
```

```julia
function vfict_explicit(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta)

    nk = size(k,1)
    kmin = minimum(k)
    kmax = maximum(k)
#     dk = (kmax-kmin)/(nk-1)
    dkf = [k[2:nk]-k[1:nk-1]; k[nk]-k[nk-1]]
    dkb = [k[2]-k[1]; k[2:nk]-k[1:nk-1]]

    # for iteration
    dVf = zeros(nk) # forward difference
    dVb = zeros(nk) # backward difference
    c   = zeros(nk) # optimal consumption (from dV)

    v = v0 # new v
    V = v  # old v

    dist  = zeros(maxit); # distance between old and new value functions at each iteration    

    # MAIN LOOP
    for n in 1:maxit

        # forward difference
        dVf[1:nk-1] = (V[2:nk]-V[1:nk-1])/dk
        dVf[nk] = 0 # will never be used (as s<0)
        # backward difference
        dVb[2:nk] = (V[2:nk]-V[1:nk-1])/dk
        dVb[1] = 0 # will never be used (as s>0)

        # consumption and savings with forward difference
        cf = dVf.^(-1/σ)
        sf = A.*k.^α - δ.*k - cf
        # consumption and savings with backward difference
        cb = dVb.^(-1/σ)
        sb = A.*k.^α - δ.*k - cb
        # consumption and derivative of value function at steady state (used for c only)
        c0  = A.*k.^α - δ.*k
        dV0 = c0.^(-σ)

        # dV_upwind makes a choice of forward or backward differences based on
        # the sign of the drift    
        If = sf .> 0       # below steady state
        Ib = sb .< 0       # above steady state
        I0 = (1.0.-If.-Ib) # at steady state

        # make sure the right approximations are used at the boundaries
        Ib[1] = 0 
        If[1] = 1
        Ib[nk] = 1
        If[nk] = 0

        dV_Upwind = dVf.*If + dVb.*Ib + dV0.*I0 # important to include third term    
        c = dV_Upwind.^(-1.0/σ)
        u = c.^(1.0-σ)/(1.0-σ)

        # CONSTRUCT MATRIX
        X = -sb.*Ib/dk
        Y = -sf.*If/dk + sb.*Ib/dk
        Z = sf.*If/dk

        A1 = spdiagm(nk,nk,[0 => Y[1:nk]]...) + spdiagm(nk,nk,[-1 => X[2:nk]]...) + spdiagm(nk,nk,[1 => Z[1:nk-1]]...)

        # explicit method, with sparse matrix
        v = Delta*(u + A1*V - ρ*V) + V # new v
        Vchange = V - v
        V = v
        
        dist[n] = maximum(abs.(Vchange))
        println([n,dist[n]])
        if dist[n]<crit
            println("Value Function Converged, Iteration = ")
            println(n)
            break
        end
    end
    
    return v, c
    
end
```

```julia
function vfict_implicit(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta)

    nk = size(k,1)
    kmin = minimum(k)
    kmax = maximum(k)
#     dk = (kmax-kmin)/(nk-1)
    dkf = [k[2:nk]-k[1:nk-1]; k[nk]-k[nk-1]]
    dkb = [k[2]-k[1]; k[2:nk]-k[1:nk-1]]

    # for iteration
    dVf = zeros(nk) # forward difference
    dVb = zeros(nk) # backward difference
    c   = zeros(nk) # optimal consumption (from dV)

    v = v0 # new v
    V = v  # old v

    dist  = zeros(maxit); # distance between old and new value functions at each iteration    

    # MAIN LOOP
    for n in 1:maxit

        # forward difference
        dVf[1:nk-1] = (V[2:nk]-V[1:nk-1])./dkf[1:nk-1]
        dVf[nk] = (A.*kmax.^α - δ.*kmax)^(-σ) # state constraint, for stability with implicit method?
#         dVf[nk] = 0 # will never be used (as s<0)
        # backward difference
        dVb[2:nk] = (V[2:nk]-V[1:nk-1])./dkb[2:nk]
        dVb[1] = (A.*kmin.^α - δ.*kmin)^(-σ) # state constraint, for stability with implicit method?
#         dVb[1] = 0 # will never be used (as s>0)

        # consumption and savings with forward difference
        cf = dVf.^(-1/σ)
        sf = A.*k.^α - δ.*k - cf
        # consumption and savings with backward difference
        cb = dVb.^(-1/σ)
        sb = A.*k.^α - δ.*k - cb
        # consumption and derivative of value function at steady state (used for c only)
        c0  = A.*k.^α - δ.*k
        dV0 = c0.^(-σ)

        # dV_upwind makes a choice of forward or backward differences based on
        # the sign of the drift    
        If = sf .> 0       # below steady state
        Ib = sb .< 0       # above steady state
        I0 = (1.0.-If.-Ib) # at steady state

#         # make sure the right approximations are used at the boundaries
#         Ib[1] = 0 
#         If[1] = 1
#         Ib[nk] = 1
#         If[nk] = 0

        dV_Upwind = dVf.*If + dVb.*Ib + dV0.*I0 # important to include third term    
        c = dV_Upwind.^(-1.0/σ)
        u = c.^(1.0-σ)/(1.0-σ)

        # CONSTRUCT MATRIX
        X = -sb.*Ib./dkb
        Y = -sf.*If./dkf + sb.*Ib./dkb
        Z = sf.*If./dkf

        A1 = spdiagm(nk,nk,[0 => Y[1:nk]]...) + spdiagm(nk,nk,[-1 => X[2:nk]]...) + spdiagm(nk,nk,[1 => Z[1:nk-1]]...)

        # implicit method, with sparse matrix
        B = (Delta*ρ + 1.0)*sparse(1.0I, nk, nk) - Delta*A1
        b = Delta*u + V
        v = B\b # new v
        Vchange = V - v
        V = v
        
        dist[n] = maximum(abs.(Vchange))
        println([n,dist[n]])
        if dist[n]<crit
            println("Value Function Converged, Iteration = ")
            println(n)
            break
        end
    end
    
    return v, c
    
end
```

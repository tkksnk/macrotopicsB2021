# Introduction

- Numerical dynamic programming in continuous time is useful and easy.


- We can sidestep root-finding problems (if any).


- We can exploit a sparsity in the approximation scheme so as to increase accuracy and efficiency of the solution


- Barles-Souganidis (1990) Theorem guarantees the uniqueness of the numerical solution.

- My lecture is based on Soyoung Lee's (Bank of Canada) lecture note and Ben Moll's (LSE) MATLAB code.

## Finite horizon

- Let's consider an optimal control problem over time period $t\in[0,T]$

$$
  v(x_{0},0) \equiv \max_{\alpha(s)} \int_{0}^{T} r(s,x(s),\alpha(s)) ds + g(x(T))
$$
subject to

$$
  \dot{x}(s) = f(s,x(s),\alpha(s)), \quad x_{0} \text{ is given}
$$
where

- $x(t) \in X \subseteq \mathbb{R}^{n}$: state vector $(n \times 1)$ at period $t$
- $\alpha(t) \in A \subseteq \mathbb{R}^{m}$: control vector $(m \times 1)$ at period $t$
- $v: X{\times}[0,T] \rightarrow \mathbb{R}$: value function
- $r: [0,T]{\times}X{\times}A \rightarrow \mathbb{R}$: instantaneous return
- $f: [0,T]{\times}X{\times}A \rightarrow X$: law of motion
- $g: X \rightarrow \mathbb{R}$: terminal value

## Example: Life-cycle model

$$
  v(a_{0},0) \equiv \max_{c(s)} \int_{0}^{T} e^{-\rho s}u(c(s)) ds
$$
subject to

$$
  \dot{a}(s) = r(s)a(s) + w(s) - c(s), \quad a_{0} \text{ is given}
$$

where
- $a$: asset
- $c$: consumption
- $u(c)$: utility
- $r$: interest rate
- $w$: wages

- We look at the problem at $t$

$$
  v(x(t),t) = \max_{\alpha(s)} \int_{t}^{T} r(s,x(s),\alpha(s)) ds + g(x(T))
$$
subject to

$$
  \dot{x}(s) = f(s,x(s),\alpha(s)), \quad x(t) \text{ is given}
$$



- Now we focus on the interval between $t$ and $t+h$

\begin{align*}
  v(x(t),t) &= \max_{\alpha(s)} \int_{t}^{t+h} r(s,x(s),\alpha(s)) ds + \int_{t+h}^{T} r(s,x(s),\alpha(s)) ds, \\
  &= \max_{\alpha(s)} \int_{t}^{t+h} r(s,x(s),\alpha(s)) ds + v(x(t+h),t+h)
\end{align*}

subject to

\begin{align*}
  \dot{x}(s) &= f(s,x(s),\alpha(s)), \\
  v(x(T),T) &= g(x(T)),
\end{align*}

$x(t)$ is given

- **Principle of Optimality**: Any continuation of the optimal plan is optimal.

## Deriving HJB

- Rearrange

$$
  0 = \max_{\alpha(s)} \int_{t}^{t+h} r(s,x(s),\alpha(s)) ds + v(x(t+h),t+h) - v(x(t),t)
$$

- Divide by $h$

$$
  0 = \max_{\alpha(s)} \frac{1}{h}\int_{t}^{t+h} r(s,x(s),\alpha(s)) ds + \color{blue}{\frac{1}{h}\big(v(x(t+h),t+h) - v(x(t),t)\big)}
$$

- As $h \rightarrow 0$

$$
  0 = \max_{\alpha(t)} r(t,x(t),\alpha(t)) + \color{blue}{v_{x}(x(t),t)\dot{x}(t) + v_{t}(x(t),t)}
$$

Note that $\frac{dv(x(t),t)}{dt} = \frac{dv}{dx}\frac{dx}{dt} + \frac{dv}{dt}$

As $\dot{x}(t)=f(t,x(t),\alpha(t))$,

$$
  0 = \max_{\alpha(t)} r(t,x(t),\alpha(t)) + v_{x}(x(t),t)f(t,x(t),\alpha(t)) + v_{t}(x(t),t)
$$

Or, denote $x(t)=x$ and $\alpha(t)=\alpha$

$$
  0 = \max_{\alpha} r(t,x,\alpha) + v_{x}(x,t)f(t,x,\alpha) + v_{t}(x,t)
$$

This is the **Hamiltonian-Jacobi-Bellman** (HJB) equation

- The HJB equation is a differential equation
- $v(x,t)$ is the solution to the HJB
- $\alpha^{*}(x,t)$ is the maximizer of the value

# Infinite horizon with discounting

$$
  v(x_{0},0) \equiv \max_{\alpha(s)} \int_{0}^{\infty} e^{-\rho s} r(x(s),\alpha(s)) ds
$$

subject to

$$
  \dot{x}(s) = f(x(s),\alpha(s)), \quad x_{0} \text{ is given.}
$$

Assume that $r$ and $f$ are not functions of $t$. (A terminal condition is also needed, which is omitted here.)

In time $t$,

$$
  v(x,t) = \max_{\alpha(s)} \int_{t}^{\infty} e^{-\rho s} r(x(s),\alpha(s)) ds
$$

subject to

$$
  \dot{x}(s) = f(x(s),\alpha(s)) \quad x=x(t) \text{ is given.}
$$

- Note that $v(x,t)$ is the value from the point of view of time 0. That is, by defining $v(x) = v(x,0)$, we have
\begin{align*}
  v(x,t) &= \max_{\alpha(s)} \int_{t}^{\infty} e^{-\rho s} r(x(s),\alpha(s)) ds, \\
  &= \max_{\alpha(s+t)} \int_{0}^{\infty} e^{-\rho (s+t)} r(x(s+t),\alpha(s+t)) ds, \\
  &= e^{-\rho t} \max_{\alpha(s+t)} \int_{0}^{\infty} e^{-\rho s} r(x(s+t),\alpha(s+t)) ds, \\
  &= e^{-\rho t} v(x)
\end{align*}

And
\begin{align*}
  v_{x}(x,t) &= \frac{\partial v(x,t)}{\partial x} = e^{-\rho t} v_{x}(x), \\
  v_{t}(x,t) &= \frac{\partial v(x,t)}{\partial t} = -\rho e^{-\rho t} v(x)
\end{align*}

- Note that $r(t,x,\alpha)=e^{-\rho t}r(x,\alpha)$ and $f(t,x,\alpha)=f(x,\alpha)$, and plug these into a HJB equation,

\begin{align*}
  0 &= \max_{\alpha} r(t,x,\alpha) + v_{x}(x,t)f(t,x,\alpha) + v_{t}(x,t), \\
  0 &= \max_{\alpha} e^{-\rho t}r(x,\alpha) + e^{-\rho t}v_{x}(x)f(x,\alpha) -\rho e^{-\rho t} v(x), \\
  0 &= \max_{\alpha} r(x,\alpha) + v_{x}(x)f(x,\alpha) -\rho v(x), \\
  \rho v(x) &= \max_{\alpha} r(x,\alpha) + v_{x}(x)f(x,\alpha)
\end{align*}

This does not depend on $t$.

## Example: Neoclassical growth model

$$
  v(k_{0}) \equiv \max_{c(t)} \int_{0}^{\infty} e^{-\rho t}u(c(t)) dt
$$

subject to

$$
  s(t) = \dot{k}(t) = f(k(t)) - \delta k(t) - c(t) \quad k_{0} \text{ is given}
$$

where $k$: capital, $s = \dot{k}$: saving, $\delta$: depreciation rate, $c$: consumption, $u(c)$: utility, and $\rho$: discount rate

- HJB equation is given by

$$
  \rho v(k) = \max_{c} u(c) + v_{k}(k)(f(k) - \delta k - c)
$$

# Numerical Solution with Finite Difference Method

- We solve on discretized grids: $\mathbf{K} = \{k_{1},...,k_{n_{k}}\}$

- Then we have

$$
  \rho v(k_{i}) = \max_{c} u(c) + v_{k}(k_{i})(f(k_{i}) - \delta k_{i} - c)
$$

- From FOC, $c = u'^{-1}(v_{k}(k))$ holds so as <b>to sidestep costly root-finding problems</b>


- How do we approximate $v_{k}(k_{i})$ at each $k_{i}\in\mathbf{K}$?



```julia
using SparseArrays
using LinearAlgebra

# NOTE 210625: the case with δ=1.0 can be solved with implicit method only?

# parameters
σ = 2.0       # CRRA
α = 0.3       # capital share
δ = 0.025 #1.0 # depreciation rate # 1.0 does not work with explicit method
ρ = 0.05      # discount rate 
A = 1         # productivity

# steady state capital
kss = (α*A/(ρ+δ))^(1/(1-α))

# number of grid points
nk = 150 # too many grid points sometimes do not work
# nk = 10001
# set up the grid
# too narrow bounds do not work (say [0.9,1.1]kss)
kmin = 0.001*kss
kmax = 2.0*kss
# evenly spaced grid
k = collect(range(kmin,stop=kmax,length=nk))
# log spaced grid
# k = collect(exp10.(range(log(kmin-kmin+1)/log(10),stop=log(kmax-kmin+1)/log(10),length=nk)))
# k = k .+ (kmin-1.0)

maxit = 10000 # maximum number of iterations
crit  = 1e-6; # convergence criteria
```

- Finite difference methods approximate $v_{k}(k_{i})$ by discretizing the derivatives with finite differences

- *Forward*: 
$$
v_{k,i}^{(F)} \equiv v_{k}^{(F)}(k_{i}) = \frac{v_{i+1}-v_{i}}{\Delta k}
$$

- *Backward*: 
$$
v_{k,i}^{(B)} \equiv v_{k}^{(B)}(k_{i}) = \frac{v_{i}-v_{i-1}}{\Delta k}
$$

where $v_{i} = v(k_{i})$ and $v_{k,i} = v_{k}(k_{i})$

Note that (graph)


- From concavity of $v(k)$, 

$$
v_{k,i}^{(F)}<v_{k,i}^{(B)}
$$
- Then, as $c=\big(v_{k}(k)\big)^{-1/\sigma}$, 

$$
c_{i}^{(F)}>c_{i}^{(B)}
$$
- Also, as $s=f(k)-{\delta}k-c$, 

$$
s_{i}^{(F)}<s_{i}^{(B)}
$$ 

- Which one to use? **Upwind scheme**

- When $s_{i}^{(F)}=f(k_{i})-\delta k_{i} - \big(v_{k,i}^{(F)}\big)^{-1/\sigma} > 0$, set $v_{k,i} = v_{k,i}^{(F)}$
<!-- $$
v_{k,i} = v_{k,i}^{(F)} = \frac{v_{i+1}-v_{i}}{\Delta k}
$$
 -->
- When $s_{i}^{(B)}=f(k_{i})-\delta k_{i} - \big(v_{k,i}^{(B)}\big)^{-1/\sigma} < 0$, set $v_{k,i} = v_{k,i}^{(B)}$ 
<!-- $$
v_{k,i} = v_{k,i}^{(B)} = \frac{v_{i}-v_{i-1}}{\Delta k}
$$ -->

- We can write a discretized HJB for each $k_{i}$

\begin{align*}
  \rho v(k_{i}) &= u(c_{i}) + v_{k}(k_{i})s_{i}, \\
  &= u(c_{i}) + v_{k,i}^{(F)} s_{i}^{(F)} \mathbb{1}_{s_{i}^{(F)}>0} + v_{k,i}^{(B)} s_{i}^{(B)} \mathbb{1}_{s_{i}^{(B)}<0} \tag{1}
\end{align*}

where

$$
  v_{k,i}^{(F)} = \frac{v_{i+1}-v_{i}}{\Delta k}, \quad v_{k,i}^{(B)} = \frac{v_{i}-v_{i-1}}{\Delta k}
$$

- Be careful of <font color="blue">the boundary conditions, $v_{k,1}^{(B)}$ and $v_{k,n_{k}}^{(F)}$</font>.

The optimal level of consumption is set to

$$
  c_{i} = u'^{-1}\big( v_{k,i}^{(F)}\mathbb{1}_{s_{i}^{(F)}>0} + v_{k,i}^{(B)}\mathbb{1}_{s_{i}^{(B)}<0} + \color{blue}{\bar{v}_{k,i}\mathbb{1}_{s_{i}^{(F)}\leq0}\mathbb{1}_{s_{i}^{(B)}\geq0}}\big)
$$

where <font color="blue">$\bar{v}_{k,i}=u'(f(k_{i}) - \delta k_{i})$ (steady state consumption)</font>

```julia
function vfict_explicit_nomatrix(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta)

    nk = size(k,1)
    kmin = minimum(k)
    kmax = maximum(k)
    # distance between each grid
    # evenly spaced grid
    dk = (kmax-kmin)/(nk-1)
    # log spaced grid
    dkf = [k[2:nk]-k[1:nk-1]; k[nk]-k[nk-1]]
    dkb = [k[2]-k[1]; k[2:nk]-k[1:nk-1]]

    # for iteration
    dVf = zeros(nk) # forward difference
    dVb = zeros(nk) # backward difference
    c   = zeros(nk) # optimal consumption (from dV)

    v = v0 # new v
    V = v  # old v

    dist  = zeros(maxit); # distance between old and new value functions at each iteration    

    # MAIN LOOP
    for n in 1:maxit

        # forward difference
        dVf[1:nk-1] = (V[2:nk]-V[1:nk-1])./dkf[1:nk-1]
        dVf[nk] = 0 # will never be used (as s<0)
        # backward difference
        dVb[2:nk] = (V[2:nk]-V[1:nk-1])./dkb[2:nk]
        dVb[1] = 0 # will never be used (as s>0)

        # consumption and savings with forward difference
        cf = dVf.^(-1/σ)
        sf = A.*k.^α - δ.*k - cf
        # consumption and savings with backward difference
        cb = dVb.^(-1/σ)
        sb = A.*k.^α - δ.*k - cb
        # consumption and derivative of value function at steady state (used for c only)
        c0  = A.*k.^α - δ.*k
        dV0 = c0.^(-σ)

        # dV_upwind makes a choice of forward or backward differences based on
        # the sign of the drift    
        If = sf .> 0       # below steady state
        Ib = sb .< 0       # above steady state
        I0 = (1.0.-If.-Ib) # at steady state

        # make sure the right approximations are used at the boundaries
        Ib[1] = 0 
        If[1] = 1
        Ib[nk] = 1
        If[nk] = 0

        dV_Upwind = dVf.*If + dVb.*Ib + dV0.*I0 # important to include third term    
        c = dV_Upwind.^(-1.0/σ)
        u = c.^(1.0-σ)/(1.0-σ)

        # explicit method, without sparse matrix
#         Vchange = u + dV_Upwind.*(A.*k.^α - δ.*k - c) - ρ.*V
        Vchange = u + (dVf.*If + dVb.*Ib).*(A.*k.^α - δ.*k - c) - ρ.*V
        v = V + Delta*Vchange # new v
        V = v

        dist[n] = maximum(abs.(Vchange))
        println([n,dist[n]])
        if dist[n]<crit
            println("Value Function Converged, Iteration = ")
            println(n)
            break
        end
    end
    
    return v, c
    
end
```

## Matrix form

By substituting $v_{k,i}^{(F)}$ and $v_{k,i}^{(B)}$ into (1), we have

\begin{align*}
  \rho v(k_{i}) &= u(c_{i}) + v_{k,i}^{(F)} s_{i}^{(F)} \mathbb{1}_{s_{i}^{(F)}>0} + v_{k,i}^{(B)} s_{i}^{(B)} \mathbb{1}_{s_{i}^{(B)}<0}, \\
  &= u(c_{i}) + \frac{v_{i+1}-v_{i}}{\Delta k} s_{i}^{(F)} \mathbb{1}_{s_{i}^{(F)}>0} + \frac{v_{i}-v_{i-1}}{\Delta k} s_{i}^{(B)} \mathbb{1}_{s_{i}^{(B)}<0}, \\
  &= u(c_{i}) + \alpha_{i} v_{i-1} + \beta_{i} v_{i} + \xi_{i} v_{i+1}
\end{align*}

where

$$
  \alpha_{i} = \frac{-s_{i}^{(B)} \mathbb{1}_{s_{i}^{(B)}<0}}{\Delta k}, \quad \beta_{i} = \frac{-s_{i}^{(F)} \mathbb{1}_{s_{i}^{(F)}>0} + s_{i}^{(B)} \mathbb{1}_{s_{i}^{(B)}<0}}{\Delta k}, \quad \xi_{i} = \frac{s_{i}^{(F)} \mathbb{1}_{s_{i}^{(F)}>0}}{\Delta k}
$$

It can be stacked into a matrix form

$$
    \rho \left[
    \begin{array}{c}
    v_{1}\\
    v_{2}\\
    v_{3}\\
    \vdots\\
    v_{i}\\
    \vdots\\
    v_{n-1}\\
    v_{n}
    \end{array} \right] = 
    \left[
    \begin{array}{c}
    u(c_{1})\\
    u(c_{2})\\
    u(c_{3})\\
    \vdots\\
    u(c_{i})\\
    \vdots\\
    u(c_{n-1})\\
    u(c_{n})
    \end{array} \right] +
    \left[
    \begin{array}{cccccccc}
    \beta_{1} & \xi_{1} & 0 & \cdots & \cdots & \cdots & \cdots & 0\\
    \alpha_{2} & \beta_{2} & \xi_{2} & 0 & \cdots & \cdots & \cdots & 0 \\
    0 & \alpha_{3} & \beta_{3} & \xi_{3} & 0 & \cdots & \cdots & 0 \\
    \vdots & & \ddots & \ddots & \ddots & & & \vdots \\
    \vdots & & 0 & \alpha_{i} & \beta_{i} & \xi_{i} & 0 & \vdots \\
    \vdots & & & & \ddots & \ddots & \ddots & \vdots \\
    0 & & & & 0 & \alpha_{n-1} & \beta_{n-1} & \xi_{n-1} \\
    0 & \cdots & \cdots & \cdots & \cdots & 0 & \alpha_{n} & \beta_{n}
    \end{array}
    \right]
    \left[
    \begin{array}{c}
    v_{1}\\
    v_{2}\\
    v_{3}\\
    \vdots\\
    v_{i}\\
    \vdots\\
    v_{n-1}\\
    v_{n}
    \end{array} \right]
$$

Or

$$
  \rho \mathbf{v} = \mathbf{u} + \mathbf{A}\mathbf{v}
$$

$\mathbf{A}$ is <b>a sparse matrix</b> (graph).

- We solve for $\mathbf{v}$ by an iterative scheme.

- Let $\mathbf{v}^{n}$ be the value function from $n$th iteration
    - **Explicit method**:

    $$
      \frac{\mathbf{v}^{n+1}-\mathbf{v}^{n}}{\Delta} + \rho \mathbf{v}^{n} = \mathbf{u}^{n} + \mathbf{A}^{n}\mathbf{v}^{n}
    $$

    - **Implicit method**:

    $$
      \frac{\mathbf{v}^{n+1}-\mathbf{v}^{n}}{\Delta} + \rho \mathbf{v}^{n+1} = \mathbf{u}^{n} + \mathbf{A}^{n}\mathbf{v}^{n+1}
    $$

where $\Delta>0$ controls the speed of updating. 

- With the implicit method, we update $\mathbf{v}^{n}$ by

\begin{align*}
  & \mathbf{v}^{n+1} + \Delta\rho\mathbf{v}^{n+1} - \Delta\mathbf{A}^{n}\mathbf{v}^{n+1} = \Delta\mathbf{u}^{n} + \mathbf{v}^{n}, \\
  \Leftrightarrow & \mathbf{v}^{n+1} = (\mathbf{I} + \Delta\rho\mathbf{I} - \Delta\mathbf{A}^{n})^{-1}(\Delta\mathbf{u}^{n} + \mathbf{v}^{n}) \\
\end{align*}


```julia
function vfict_explicit(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta)

    nk = size(k,1)
    kmin = minimum(k)
    kmax = maximum(k)
    dk = (kmax-kmin)/(nk-1)
#     dkf = [k[2:nk]-k[1:nk-1]; k[nk]-k[nk-1]]
#     dkb = [k[2]-k[1]; k[2:nk]-k[1:nk-1]]

    # for iteration
    dVf = zeros(nk) # forward difference
    dVb = zeros(nk) # backward difference
    c   = zeros(nk) # optimal consumption (from dV)

    v = v0 # new v
    V = v  # old v

    dist  = zeros(maxit); # distance between old and new value functions at each iteration    

    # MAIN LOOP
    for n in 1:maxit

        # forward difference
        dVf[1:nk-1] = (V[2:nk]-V[1:nk-1])/dk
        dVf[nk] = 0 # will never be used (as s<0)
        # backward difference
        dVb[2:nk] = (V[2:nk]-V[1:nk-1])/dk
        dVb[1] = 0 # will never be used (as s>0)

        # consumption and savings with forward difference
        cf = dVf.^(-1/σ)
        sf = A.*k.^α - δ.*k - cf
        # consumption and savings with backward difference
        cb = dVb.^(-1/σ)
        sb = A.*k.^α - δ.*k - cb
        # consumption and derivative of value function at steady state (used for c only)
        c0  = A.*k.^α - δ.*k
        dV0 = c0.^(-σ)

        # dV_upwind makes a choice of forward or backward differences based on
        # the sign of the drift    
        If = sf .> 0       # below steady state
        Ib = sb .< 0       # above steady state
        I0 = (1.0.-If.-Ib) # at steady state

        # make sure the right approximations are used at the boundaries
        Ib[1] = 0 
        If[1] = 1
        Ib[nk] = 1
        If[nk] = 0

        dV_Upwind = dVf.*If + dVb.*Ib + dV0.*I0 # important to include third term    
        c = dV_Upwind.^(-1.0/σ)
        u = c.^(1.0-σ)/(1.0-σ)

        # CONSTRUCT MATRIX
        X = -sb.*Ib/dk
        Y = -sf.*If/dk + sb.*Ib/dk
        Z = sf.*If/dk

        A1 = spdiagm(nk,nk,[0 => Y[1:nk]]...) + spdiagm(nk,nk,[-1 => X[2:nk]]...) + spdiagm(nk,nk,[1 => Z[1:nk-1]]...)

        # explicit method, with sparse matrix
        v = Delta*(u + A1*V - ρ*V) + V # new v
        Vchange = V - v
        V = v
        
        dist[n] = maximum(abs.(Vchange))
        println([n,dist[n]])
        if dist[n]<crit
            println("Value Function Converged, Iteration = ")
            println(n)
            break
        end
    end
    
    return v, c
    
end
```

```julia
function vfict_implicit(v0,k,A,σ,α,δ,ρ,maxit,crit,Delta)

    nk = size(k,1)
    kmin = minimum(k)
    kmax = maximum(k)
#     dk = (kmax-kmin)/(nk-1)
    dkf = [k[2:nk]-k[1:nk-1]; k[nk]-k[nk-1]]
    dkb = [k[2]-k[1]; k[2:nk]-k[1:nk-1]]

    # for iteration
    dVf = zeros(nk) # forward difference
    dVb = zeros(nk) # backward difference
    c   = zeros(nk) # optimal consumption (from dV)

    v = v0 # new v
    V = v  # old v

    dist  = zeros(maxit); # distance between old and new value functions at each iteration    

    # MAIN LOOP
    for n in 1:maxit

        # forward difference
        dVf[1:nk-1] = (V[2:nk]-V[1:nk-1])./dkf[1:nk-1]
        dVf[nk] = (A.*kmax.^α - δ.*kmax)^(-σ) # state constraint, for stability with implicit method?
#         dVf[nk] = 0 # will never be used (as s<0)
        # backward difference
        dVb[2:nk] = (V[2:nk]-V[1:nk-1])./dkb[2:nk]
        dVb[1] = (A.*kmin.^α - δ.*kmin)^(-σ) # state constraint, for stability with implicit method?
#         dVb[1] = 0 # will never be used (as s>0)

        # consumption and savings with forward difference
        cf = dVf.^(-1/σ)
        sf = A.*k.^α - δ.*k - cf
        # consumption and savings with backward difference
        cb = dVb.^(-1/σ)
        sb = A.*k.^α - δ.*k - cb
        # consumption and derivative of value function at steady state (used for c only)
        c0  = A.*k.^α - δ.*k
        dV0 = c0.^(-σ)

        # dV_upwind makes a choice of forward or backward differences based on
        # the sign of the drift    
        If = sf .> 0       # below steady state
        Ib = sb .< 0       # above steady state
        I0 = (1.0.-If.-Ib) # at steady state

#         # make sure the right approximations are used at the boundaries
#         Ib[1] = 0 
#         If[1] = 1
#         Ib[nk] = 1
#         If[nk] = 0

        dV_Upwind = dVf.*If + dVb.*Ib + dV0.*I0 # important to include third term    
        c = dV_Upwind.^(-1.0/σ)
        u = c.^(1.0-σ)/(1.0-σ)

        # CONSTRUCT MATRIX
        X = -sb.*Ib./dkb
        Y = -sf.*If./dkf + sb.*Ib./dkb
        Z = sf.*If./dkf

        A1 = spdiagm(nk,nk,[0 => Y[1:nk]]...) + spdiagm(nk,nk,[-1 => X[2:nk]]...) + spdiagm(nk,nk,[1 => Z[1:nk-1]]...)

        # implicit method, with sparse matrix
        B = (Delta*ρ + 1.0)*sparse(1.0I, nk, nk) - Delta*A1
        b = Delta*u + V
        v = B\b # new v
        Vchange = V - v
        V = v
        
        dist[n] = maximum(abs.(Vchange))
        println([n,dist[n]])
        if dist[n]<crit
            println("Value Function Converged, Iteration = ")
            println(n)
            break
        end
    end
    
    return v, c
    
end
```

# Barles-Souganidis

- We use the upwind scheme to satisfy one of the conditions that are required to apply Barles-Souganidis (1990) Theorem:

If the scheme satisfies the monotonicity, consistency and stability conditions, then as $\Delta k \rightarrow 0$ its solution $v_{i}, i=1,...,n$ converges locally uniformly to the uniquely viscosity solution of $G$

- Can write any HJB equation with one state variable as

$$
  0 = G(k,v(k),v_{k}(k)) = \rho v(k) - \max_{c} u(c) - v_{k}(k)(f(k) - \delta k - c)
$$

- Corresponding numerical scheme

\begin{align*}
  0 &= S(\Delta k,k_{i},v_{i},v_{i-1},v_{i+1}) \\
  &= \rho v(k_{i}) - u(c_{i}) - \frac{v_{i+1}-v_{i}}{\Delta k}(f(k_{i})-\delta k_{i}-c_{i})^{+} - \frac{v_{i}-v_{i-1}}{\Delta k}(f(k_{i})-\delta k_{i}-c_{i})^{-}
\end{align*}

1. Monotonicity: the numerical scheme is monotone, that is $S$ is non-increasing in both $v_{i-1}$ and $v_{i+1}$.


2. Consistency: the numerical scheme is consistent, that is for every smooth function $v$ with bounded derivatives, 
$$
  S(\Delta k,k_{i},v_{i},v_{i-1},v_{i+1}) \rightarrow G(k,v(k),v_{k}(k))
$$
as $\Delta k \rightarrow 0$ and $k_{i} \rightarrow k$.

3. Stability: the numerical scheme is stable, that is for every $\Delta k > 0$, it has a solution $v_{i}, i=1,...,n$ which is uniformly bounded independently of $\Delta k$.

- Check monotonicity of the scheme. Recall

\begin{align*}
  & S(\Delta k,k_{i},v_{i},v_{i-1},v_{i+1}) \\
  &= \rho v(k_{i}) - u(c_{i}) - \frac{v_{i+1}-v_{i}}{\Delta k}\dot{k}^{+} - \frac{v_{i}-v_{i-1}}{\Delta k}\dot{k}^{-} \\
  &= \rho v(k_{i}) - u(c_{i}) - \frac{\dot{k}^{+}}{\Delta k}v_{i+1} + \frac{\dot{k}^{+}}{\Delta k}v_{i} - \frac{\dot{k}^{-}}{\Delta k}v_{i} + \frac{\dot{k}^{-}}{\Delta k}v_{i-1}  \\
\end{align*}

- $S(\Delta k,k_{i},v_{i},v_{i-1},v_{i+1})$ is non-increasing in $v_{i-1}$ and $v_{i+1}$.
